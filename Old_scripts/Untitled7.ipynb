```python
import numpy as np
def z_score_normalization(image_data):
    mean = np.mean(image_data)
    std = np.std(image_data)
    
    
    normalized_image = (image_data - mean) / std
    
    return normalized_image
```

```python
import os
import nibabel as nib

patient_folders = sorted(os.listdir(data_dir))[:40]  
modality_keys = ["flair", "t1", "t1ce", "t2"]

patient_data_dict = {}


for patient in patient_folders:
    patient_path = os.path.join(data_dir, patient)
    
    print(f"Loading patient: {patient}")
    
    
    patient_modalities = {}
    
    for modality in modality_keys:
        filename = f"{patient}_{modality}.nii"
        file_path = os.path.join(patient_path, filename)
        
        if os.path.exists(file_path):
            
            img = nib.load(file_path)
            data = img.get_fdata()
            print(f"  {modality.upper()} loaded - shape: {data.shape}")
            
            
            patient_modalities[modality] = data
        else:
            print(f"  Missing {modality.upper()} - {filename}")
    
    
    patient_data_dict[patient] = patient_modalities



```

```python
for patient, modalities in patient_data_dict.items():
    print(f"Normalizing patient: {patient}")
    
    
    for modality, image_data in modalities.items():
        
        normalized_image = z_score_normalization(image_data)
        
        
        patient_data_dict[patient][modality] = normalized_image
        
print("Complete")


```

```python
def min_max_scaling(image_data):
 
    min_val = np.min(image_data)
    max_val = np.max(image_data)
    
    scaled_image = (image_data - min_val) / (max_val - min_val)
    
    return scaled_image
```

```python
for patient, modalities in patient_data_dict.items():
    for modality, image_data in modalities.items():
        
        scaled_data = min_max_scaling(image_data)
        
        
        patient_data_dict[patient][modality] = scaled_data
print("Complete")
```
